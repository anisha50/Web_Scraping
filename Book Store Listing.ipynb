```python
import requests
from bs4 import BeautifulSoup
```

```python
pages=[]
for i in range(1, 51):
    url = 'http://books.toscrape.com/catalogue/page-' + str(i) + '.html'
    pages.append(url)

pages
```

```python
URL=[]
Title=[]
Price=[]
Rating=[]
Stock=[]
Image=[]
for item in pages:
    page=requests.get(item)
    soup = BeautifulSoup(page.text, 'html.parser')
    book_name_list = soup.find_all(class_='product_pod')
    for books in book_name_list:
        name=books.find_all('a')
        price=books.find(class_='product_price')
        stock=books.find(class_='instock availability').get_text()
        image=books.find(class_='thumbnail').get('src')
        if bool(books.find(class_='star-rating One')):
            rating=1
        if bool(books.find(class_='star-rating Two')):
            rating=2
        if bool(books.find(class_='star-rating Three')):
            rating=3
        if bool(books.find(class_='star-rating Four')):
            rating=4
        if bool(books.find(class_='star-rating Five')):
            rating=5  
        if stock=='\n\n    \n        In stock\n    \n':
            status=True
        else:
            status=False
        URL.append(name[1].get('href'))
        Title.append(name[1].get('title'))
        Price.append(price.find(class_='price_color').get_text()[1:])
        Rating.append(rating)
        Stock.append(status)
        Image.append(image)
    
```

```python


import pandas as pd
df=pd.DataFrame(Title,columns=['Title'])
df['Rating']=Rating
df['Stock']=Stock
df['Cover']=Image
df['URL']=URL
df['Price']=Price
df
```
